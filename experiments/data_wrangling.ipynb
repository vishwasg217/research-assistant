```python
import pandas as pd

# import json to pandas
df = pd.read_json("../data/aiml_papers.json")

df[:5]
```

```python
import json


with open("../data/paper_categories.json", "r") as f:
    paper_categories = json.load(f)

def get_category_names(cat_str):
    cat_list = cat_str.split(' ')
    return [paper_categories.get(cat, cat) for cat in cat_list]

df['categories'] = df['categories'].apply(get_category_names)

df[:5]
```

```python
def format_authors(authors: list[list]): 
    return [' '.join(author).strip() for author in authors]

df['authors'] = df['authors_parsed'].apply(format_authors)

list(df['authors'][:5])
```

```python
df['update_date'] = pd.to_datetime(df['update_date']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
df['create_date'] = pd.to_datetime(df['versions'].apply(lambda x: x[0]['created'])).dt.strftime('%Y-%m-%dT%H:%M:%SZ')

print(df['update_date'][:5])
print(df['create_date'][:5])
```
